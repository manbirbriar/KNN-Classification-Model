```python
knn = KNeighborsClassifier(n_neighbors=3)
knn=knn.fit(irisX_train, sy_train)

x = 1
k_values = []
train_accuracy = []
test_accuracy = []
while x <= 5:
    knn = KNeighborsClassifier(n_neighbors=x)
    knn.fit(x_train, y_train)
    k_values.append(x)
    train_accuracy.append(knn.score(x_train, y_train))
    test_accuracy.append(knn.score(x_test, y_test))
    x += 2

```

```python
fig, ax = plt.subplots()
plt.scatter(k_values, train_accuracy, color='blue', label='Training vs. K Value')
plt.xlabel('K Value')
plt.ylabel('Training Dataset Accuracy')
plt.legend(loc='upper right')
plt.show()
```

```python

fig, ax = plt.subplots()
plt.scatter(k_values, test_accuracy, color='blue', label='Test vs. K Value')
plt.xlabel('K Value')
plt.ylabel('Test Dataset Accuracy')
plt.legend(loc='upper right')
plt.show()
```
